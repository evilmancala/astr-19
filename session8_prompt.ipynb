Create a Jupyter notebook, import matplotlib. 

Write cells that: 

1) create a numpy array x ranging from [0,1] in 100 steps and 

2) that defines functions that return sin(x) and cos(x).

3) In a new cell use matplotlib to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 100
x = np.linspace(0, 1, n)

sinx = np.sin(x)
cosx = np.cos(x)
```

Plotting panels. Emphasis on the plural.

```python
figure, axes = plt.subplots(1,2)

axes[0].plot(x,sinx)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('sin(x)')
axes[0].set_xlim(np.min(x),np.max(x))
axes[0].set_ylim(np.min(sinx),np.max(sinx))

#second panel
axes[1].plot(x,cosx)
axes[1].set_xlabel('x')
#axes[1].set_ylabel('y')
axes[1].set_title('cos(x)')
axes[1].set_xlim(np.min(x),np.max(x))
axes[1].set_ylim(np.min(sinx),np.max(sinx))

axes[1].set_yticklabels([])
axes[1].tick_params(axis='y',direction='inout')
figure.subplots_adjust(wspace=0.05)

plt.savefig("session-8-prompt.pdf",bbox_inches="tight",dpi=300,facecolor='white')
```
