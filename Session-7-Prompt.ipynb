## Prompt 

Create a Jupyter notebook, import matplotlib. 

Write cells that 

1) create a numpy array x ranging from [0,1] in 100 steps and 

2) that defines a function that returns exp(x).

3) In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF.  

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Array and Function

```python
#Defining data
n = 100
x = np.linspace(0,1,n) #create an array of x and n points from 0 to 2*pi

#Defining how to make exp(x)
def exp(x):
    return np.sin(x) 
y = exp(x)
```

Showing the plot

```python
plt.plot(x,y)         

#Labels
plt.xlabel("Time [milliseconds]")    #add x axis label
plt.ylabel("Awesomeness")  

#saving
plt.savefig("session-7-prompt.pdf",bbox_inches="tight",dpi=300,facecolor='white')

plt.show()
```
