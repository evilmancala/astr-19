```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Choosing Distribution: von Mises
from scipy.special import i0  

#Drawing the sample from the recorded data

mu, kappa = 0.0, 4.0 #Mu: Mode (“center”) of the distribution. /// Kappa: Concentration of the distribution, has to be >=0.
n = np.random.vonmises(mu, kappa, 1000)

#taken from numpy website
```

```python
#Bins, code taken (and minorly edited) from numpy website. Collection 100x

plt.hist(n, 100, density=True)
x = np.linspace(-np.pi, np.pi, num=100) #Looked up the meaning (AI): Made to space out the values according to 
                                                                    #pi due to the circular nature of vM. 
y = np.exp(kappa*np.cos(x-mu))/(2*np.pi*i0(kappa))  #Meant to be the code version of the von Mises equation.
plt.plot(x, y, linewidth=2, color='r')  


#Labeling and Saving (the easy part!)
plt.xlabel('X Values')
plt.ylabel('Counts')
plt.savefig("session-10-prompt.pdf",bbox_inches="tight",dpi=300,facecolor='white')
plt.show()
```
